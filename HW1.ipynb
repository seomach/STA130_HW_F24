## 6.)
In `df.describe()`, we can observe that only numerical columns are summarized. That is because `describe()` is designed to provide statistical insights that are better suited for numerical data. We can use `df.describe(include='all')`, but it would return NaN (or "Not a Number") for most of the statistics in the non-numerical column.

Here is what each statistic featured in `df.describe()` means:

> count - represents the number of values present in each column, excluding null or missing values.

> mean - the calculation of the sum of all non-null values divided by the count of non-null values; represents the average value in the column and provides a measure of central tendency.

> std/standard deviation - the standard deviation is the measure of the spread of the values in the column; assessing how much the values deviate from the mean on average. High standard deviations indicate a wide spread of values, while low standard deviations indicate that the values are close to the mean.

> min - represents the smallest number or the minimum value among all non-null values in the column.

> 25% - represents the measure of the lower end of the data distribution; particularly the value below which 25% of the data falls. This term can also be known as the first quartile (Q1).

> 50% - represents the middle value of the column wherein half the values are below this value, and half are above the value. It is a measure of central tendency that is less affected by extremely high or low values (or the 'outliers') than the mean.

> 75% - represents the measure of the higher end of the data distribution; particularly the value below which 75% of the data falls. This term can also be known as the third quartile (Q3).

> max - the largest value or maximum value among all the non-null values in the column.

It can be observed that `df.describe()` ignores missing values when computing summary statistics. In the statistic *count*, for example, *count* only retrieves the number of 'present' values -- which means missing values are disregarded.

## 7.a)

We can observe that `dropna()` is preferably used when we intend to 'drop' or remove columns or rows with missing values. The `del df['col']` removes an entire column (which I assume it doesn't work on rows), regardless if it has missing values or not.

> Accuracy and validity are important aspects in Statistics. Accuracy describes the reliability of the data so that they can be used as a trustworthy source of data and/or support better decision-making, while validity ensures that the data is consistent and 'reasonable', based on defined set of rules.

Missing values can affect accuracy and validity in a dataset. So, in the case of a dataset with missing values, `dropna()` is preferred when we want to 'clean' or eliminate null entries from the dataset.

When speaking about the most efficient way to remove missing values, `dropna()` is more efficient than `del` since in `del` we might have to figure out which columns have missing values.

Another thing to take note of is that we can control which axis (rows or columns) to remove when using `dropna()`. Columns are in `axis=1` while rows are represented by `axis=0`. So, `dropna()` will be preferable over `del df['col']` in the case of wanting to remove a row with missing values -- considering that `del df['col']` only works on columns. 

___

## 7.b)
On the other hand, the `del` function is preferably used when we intend to remove a column entirely, regardless if it has missing values.

If we assert that a column is irrelevant to our analysis, we can just use `del df['col']` to remove that column. 

In the case of an irrelevant column in the dataset, `del` is more efficient because we are disregarding if there could be any missing values and more concerned about how relevant a column is.

___

## 7.c)

If we suspect that a column is irrelevant to our analysis and also has missing values, we must first use `del df['col']` before `dropna()`.

This is because if we use `dropna()` first, it would remove rows (assuming we want to remove rows and only rows using `dropna()`) that have missing values in the irrelevant column but non-null values in the relevant columns.

Executing `del df['col']` first ensures that only the relevant data that are necessary for the analysis are present, and no other instances of the data are removed when using `dropna()` -- thus preventing the removal of rows whose entries are valid in the relevant columns, but has a missing value in the irrelevant column.

```python
# Before
df
```

```python
# After
df_cleaned = df.dropna()
df_cleaned
```

```python
# We just removed the rows with missing values so 20640 rows - 207 rows equal to 20433 rows.
# No columns are removed.

# To provide more documentation

# Before | Report

shape_before = df.shape
nonan_before = df.isna().sum()
desc_before = df.describe()

print(f"Initial dataset shape: {shape_before}")
print("\nMissing values before using dropna():")
print(nonan_before)
print("\nDataset description before using dropna():")
print(desc_before)
```

```python
# After | Report

shape_after = df_cleaned.shape
nonan_after = df_cleaned.isna().sum()
desc_after = df_cleaned.describe()

print(f"New dataset shape: {shape_after}")
print("\nMissing values after using dropna():")
print(nonan_after)
print("\nDataset description after using dropna():")
print(desc_after)
```

The summaries for the report after cleaning the dataset now differ slightly from the report before cleaning the dataset. The most notable being the count, mean, and standard deviation. This is because we removed the rows with missing values and now for all the columns in the 'after' report now have a count of 20433 which can affect the mean, which can, in turn, affect the standard deviation.

## 8.a)

I will explain `df.groupby('col1')['col2'].describe()` in different parts:

First, the `df.groupby('col1')` method groups the DataFrame using the values from 'col1' and returns a grouped or GroupBy object that corresponds to a unique value in 'col1':


```python
# Let's say we want to find out how many survived based on their sex. Let's first group the dataset by 'sex'.
df_titanic.groupby('sex')
```

```python
# We can observe here that it returns a GroupBy object.

# Next, the ['col2'] in groupby('col1')['col2'].describe() indicates the specified column where subsequent operations will be performed for each group.

# Here we replace 'col2' with the column 'survived'
df_titanic.groupby('sex')['survived']
```

```python
# It still returns a GroupBy object, but with a different hexadecimal memory address

# To provide information we can actually understand, this is where describe() comes in.
# The method describe(), as demonstrated before, generates the descriptive statistics of the dataset. 
# In the context of groupby('col1')['col2'].describe(), describe() here is performed specifically on the selected column -- which is the column 'survive' -- corresponding to each unique value in 'col1'.

# To reiterate, the groupby() method takes 'col1' as an argument, then we specify which columns (enclosed in brackets -- ['col2']) we want to see be corresponded with a unique value in 'col1'. 
# After that, we use describe() to generate the descriptive statistics of the specified column, which is 'col2'.

# Using them altogether:
df_titanic.groupby('sex')['survived'].describe()

# Here I intended to generate descriptive statistics of the survival rate based on whether or not they are male or female.
```
